```python
file_name = "wdbc.csv"
open_file = open(file_name, "rU")
csv_data = csv.reader(open_file)
```

```python
# data_array[:, 1] ## get column index 1
# data_array[0, :] ## get row index 0
```

```python
X = data_array[:, 2:32].astype(float) ## => columns 2 to 32: measurements => DATA MATRIX
Y = data_array[:, 1]                  ## => column1 : malignant (M) or begnin (B) => CLASS VECTOR
```

```python
print "X dimension: ", X.shape
print "Y dimension: ", Y.shape
```

```python
Yfreq = sp.stats.itemfreq(Y)
print Yfreq
```
